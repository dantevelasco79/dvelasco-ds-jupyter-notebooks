```python
import pandas as pd

df_spam = pd.read_csv('spambase.csv')
df_spam.head()
```

```python
from sklearn.preprocessing import StandardScaler

columns_spam = df_spam.columns.tolist()
columns_spam.remove('class')

std_scaler = StandardScaler()
for column in columns_spam:
    df_spam[column] = std_scaler.fit_transform(df_spam[column].values.reshape(-1,1))
df_spam.head()
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split

X_spam = df_spam[columns_spam]
y_spam = df_spam['class']

X_spam_train,X_spam_test,y_spam_train,y_spam_test=train_test_split(X_spam,y_spam,test_size=0.25)

clf_spam = tree.DecisionTreeClassifier(criterion='entropy')
clf_spam.fit(X_spam_train, y_spam_train)
```

```python
from sklearn import metrics
from sklearn.metrics import roc_auc_score

y_spam_pred_train=clf_spam.predict(X_spam_train)
cm_spam_train = metrics.confusion_matrix(y_spam_train, y_spam_pred_train)

cm_spam_train_tp = cm_spam_train[0][0]
cm_spam_train_fp = cm_spam_train[1][0]
cm_spam_train_tn = cm_spam_train[1][1]
cm_spam_train_fn = cm_spam_train[0][1]

accuracy_train_spam = (cm_spam_train_tp + cm_spam_train_tn) / (cm_spam_train_tp + cm_spam_train_tn + cm_spam_train_fp + cm_spam_train_fn)
error_train_spam = 1 - accuracy_train_spam
precision_train_spam = cm_spam_train_tp / (cm_spam_train_tp + cm_spam_train_fp)
recall_train_spam = cm_spam_train_tp / (cm_spam_train_tp + cm_spam_train_fn)
f1_score_train_spam = 2 * (precision_train_spam * recall_train_spam)/(precision_train_spam + recall_train_spam)
auc_score_train_spam = roc_auc_score(y_spam_train, clf_spam.predict_proba(X_spam_train)[:, 1])
```

```python
y_spam_pred_test=clf_spam.predict(X_spam_test)
cm_spam_test = metrics.confusion_matrix(y_spam_test, y_spam_pred_test)

cm_spam_test_tp = cm_spam_test[0][0]
cm_spam_test_fp = cm_spam_test[1][0]
cm_spam_test_tn = cm_spam_test[1][1]
cm_spam_test_fn = cm_spam_test[0][1]

accuracy_test_spam = (cm_spam_test_tp + cm_spam_test_tn) / (cm_spam_test_tp + cm_spam_test_tn + cm_spam_test_fp + cm_spam_test_fn)
error_test_spam = 1 - accuracy_test_spam
precision_test_spam = cm_spam_test_tp / (cm_spam_test_tp + cm_spam_test_fp)
recall_test_spam = cm_spam_test_tp / (cm_spam_test_tp + cm_spam_test_fn)
f1_score_test_spam = 2 * (precision_test_spam * recall_test_spam)/(precision_test_spam + recall_test_spam)
auc_score_test_spam = roc_auc_score(y_spam_test, clf_spam.predict_proba(X_spam_test)[:, 1])
```

```python
print('For training, ERR: ' + str(error_train_spam) + ', ACC: ' + str(accuracy_train_spam) + ', F1 Score: ' + str(f1_score_train_spam) + ', AUC Score: ' + str(auc_score_train_spam))
print('For testing, ERR: ' + str(error_test_spam) + ', ACC: ' + str(accuracy_test_spam) + ', F1 Score: ' + str(f1_score_test_spam) + ', AUC Score: ' + str(auc_score_test_spam))
```

#### The metrics are all very good. This model seems to be strong.

```python
import matplotlib.pyplot as plt

def plot_depths():
    tree_depths = range(10)
    total_errors = []
    train_errors = []
    test_errors = []
    for depth in tree_depths:
        depth += 1
        
        clf_spam = tree.DecisionTreeClassifier(max_depth=depth)
        clf_spam.fit(X_spam_train, y_spam_train)
        
        y_spam_pred_train=clf_spam.predict(X_spam_train)
        cm_spam_train = metrics.confusion_matrix(y_spam_train, y_spam_pred_train)

        cm_spam_train_tp = cm_spam_train[0][0]
        cm_spam_train_fp = cm_spam_train[1][0]
        cm_spam_train_tn = cm_spam_train[1][1]
        cm_spam_train_fn = cm_spam_train[0][1]

        accuracy_train_spam = (cm_spam_train_tp + cm_spam_train_tn) / (cm_spam_train_tp + cm_spam_train_tn + cm_spam_train_fp + cm_spam_train_fn)
        error_train_spam = 1 - accuracy_train_spam
        
        y_spam_pred_test=clf_spam.predict(X_spam_test)
        cm_spam_test = metrics.confusion_matrix(y_spam_test, y_spam_pred_test)

        cm_spam_test_tp = cm_spam_test[0][0]
        cm_spam_test_fp = cm_spam_test[1][0]
        cm_spam_test_tn = cm_spam_test[1][1]
        cm_spam_test_fn = cm_spam_test[0][1]

        accuracy_test_spam = (cm_spam_test_tp + cm_spam_test_tn) / (cm_spam_test_tp + cm_spam_test_tn + cm_spam_test_fp + cm_spam_test_fn)
        error_test_spam = 1 - accuracy_test_spam
        
        total_errors.append(error_train_spam + error_test_spam)
        train_errors.append(error_train_spam)
        test_errors.append(error_test_spam)
        
    f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
    ax1.set_xlabel('Tree Depths')
    ax1.set_ylabel('Train Error')
    ax1.set_title('Best Depth Plot')
    ax1.plot(tree_depths, train_errors)
    
    ax2.set_xlabel('Tree Depths')
    ax2.set_ylabel('Test Error')
    ax2.set_title('Best Depth Plot')
    ax2.plot(tree_depths, test_errors)
    
plot_depths()
```

#### Training Error continues to decrease, but Test Error seems to plateau around depth 5, so I recommend this depth to prevent overfitting.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

def ensembles(forest):
    thresholds = [10, 50, 100, 500]
    for T in thresholds:
        
        if forest:
            clf_spam = RandomForestClassifier(n_estimators=T)
        else:
            clf_spam = AdaBoostClassifier(n_estimators=T)
        clf_spam.fit(X_spam_train, y_spam_train)
        
        y_spam_pred_train=clf_spam.predict(X_spam_train)
        cm_spam_train = metrics.confusion_matrix(y_spam_train, y_spam_pred_train)

        cm_spam_train_tp = cm_spam_train[0][0]
        cm_spam_train_fp = cm_spam_train[1][0]
        cm_spam_train_tn = cm_spam_train[1][1]
        cm_spam_train_fn = cm_spam_train[0][1]

        accuracy_train_spam = (cm_spam_train_tp + cm_spam_train_tn) / (cm_spam_train_tp + cm_spam_train_tn + cm_spam_train_fp + cm_spam_train_fn)
        error_train_spam = 1 - accuracy_train_spam
        precision_train_spam = cm_spam_train_tp / (cm_spam_train_tp + cm_spam_train_fp)
        recall_train_spam = cm_spam_train_tp / (cm_spam_train_tp + cm_spam_train_fn)
        f1_score_train_spam = 2 * (precision_train_spam * recall_train_spam)/(precision_train_spam + recall_train_spam)
        auc_score_train_spam = roc_auc_score(y_spam_train, clf_spam.predict_proba(X_spam_train)[:, 1])
        
        y_spam_pred_test=clf_spam.predict(X_spam_test)
        cm_spam_test = metrics.confusion_matrix(y_spam_test, y_spam_pred_test)

        cm_spam_test_tp = cm_spam_test[0][0]
        cm_spam_test_fp = cm_spam_test[1][0]
        cm_spam_test_tn = cm_spam_test[1][1]
        cm_spam_test_fn = cm_spam_test[0][1]

        accuracy_test_spam = (cm_spam_test_tp + cm_spam_test_tn) / (cm_spam_test_tp + cm_spam_test_tn + cm_spam_test_fp + cm_spam_test_fn)
        error_test_spam = 1 - accuracy_test_spam
        precision_test_spam = cm_spam_test_tp / (cm_spam_test_tp + cm_spam_test_fp)
        recall_test_spam = cm_spam_test_tp / (cm_spam_test_tp + cm_spam_test_fn)
        f1_score_test_spam = 2 * (precision_test_spam * recall_test_spam)/(precision_test_spam + recall_test_spam)
        auc_score_test_spam = roc_auc_score(y_spam_test, clf_spam.predict_proba(X_spam_test)[:, 1])
        
        print('For training with T: '+ str(T) + ', ERR: ' + str(error_train_spam) + ', ACC: ' + str(accuracy_train_spam) + ', F1 Score: ' + str(f1_score_train_spam) + ', AUC Score: ' + str(auc_score_train_spam))
        print('For testing with T: ' + str(T) + ', ERR: ' + str(error_test_spam) + ', ACC: ' + str(accuracy_test_spam) + ', F1 Score: ' + str(f1_score_test_spam) + ', AUC Score: ' + str(auc_score_test_spam))
    
ensembles(True)
```

#### The metrics get marginally better for testing as T increases, but eventually they plateau. For example, the difference in accuracies between thresholds 10 and 50 is about .005, the difference in accuracies between thresholds 50 and 100 is about .001, and there is no noticeable difference in accuracy between thresholds 100 and 500.

```python
df_mush = pd.read_excel("agaricus-lepiota.xlsx")

df_mush_og = df_mush.copy()

columns_mush = df_mush.columns.tolist()
for column in columns_mush:
    df_mush[column] = pd.factorize(df_mush[column])[0]

columns_mush.remove('edibility')
    
df_mush.head()
```

```python
from sklearn.naive_bayes import MultinomialNB

X_mush = df_mush[columns_mush]
y_mush = df_mush['edibility']

X_mush_train,X_mush_test,y_mush_train,y_mush_test=train_test_split(X_mush,y_mush,test_size=0.25)

clf_mush = MultinomialNB()
clf_mush.fit(X_mush_train, y_mush_train)
```

```python
from math import e

log_prior = clf_mush.class_log_prior_
log_prob = clf_mush.feature_log_prob_

prior = []
probs = []
for log_p in log_prior:
    prior.append(e**log_p)
for p_class in log_prob:
    for log_p in p_class:
        probs.append(e**log_p)
classes = pd.unique(df_mush_og['edibility'])
for idx in range(len(classes)):
    print('The probability of class ' + classes[idx] + ' for edibility is ' + str(prior[idx]))
```

```python
feature_classes = []
for column in columns_mush:
    feature_classes.append(pd.unique(df_mush_og[column]))

calculated_probs = []
for t_class in classes:
    df_temp = df_mush_og[df_mush_og['edibility'] == t_class]
    prob_to_add = []
    for idx in range(len(columns_mush)):
        p_to_add = []
        df_counts = df_temp[columns_mush[idx]].value_counts()
        total = 8124
        num_classes = len(feature_classes[idx])
        begin = 0
        for jdx in range(num_classes):
            if feature_classes[idx][jdx] not in df_counts:
                count = 0
            else:
                count = df_counts[begin]
                begin += 1
            prob = (count + 1) / (total + num_classes)
            print('The probability of class ' + feature_classes[idx][jdx] + ' for ' + columns_mush[idx] + ' given edibility ' + t_class + ' is ' + str(prob))
            p_to_add.append(prob)
        prob_to_add.append(p_to_add)
    calculated_probs.append(prob_to_add)
```

## b)

```python
y_pred_manual = []

for idx in range(X_mush_test[columns_mush[0]].count()):
    row_values = X_mush_test.iloc[[idx][0]]
    prob_poison = 1
    prob_edible = 1
    for jdx in range(len(row_values)):
        prob_poison *= calculated_probs[0][jdx][row_values[jdx]]
        prob_edible *= calculated_probs[1][jdx][row_values[jdx]]
    if (prob_poison > prob_edible):
        y_pred_manual.append(0)
    else:
        y_pred_manual.append(1)

y_pred_manual
```

## c)

```python
cm_mush_tp_manual = 0
cm_mush_fp_manual = 0
cm_mush_tn_manual = 0
cm_mush_fn_manual = 0
    
for idx in range(len(y_pred_manual)):
    if (y_pred_manual[idx] == 1 and y_mush_test.iloc[idx] == 1):
        cm_mush_tp_manual += 1
    elif (y_pred_manual[idx] == 1 and y_mush_test.iloc[idx] == 0):
        cm_mush_fp_manual += 1
    elif (y_pred_manual[idx] == 0 and y_mush_test.iloc[idx] == 0):
        cm_mush_tn_manual += 1
    else:
        cm_mush_fn_manual += 1

accuracy_mush_manual = (cm_mush_tp_manual + cm_mush_tn_manual) / (cm_mush_tp_manual + cm_mush_tn_manual + cm_mush_fp_manual + cm_mush_fn_manual)
error_mush_manual = 1 - accuracy_mush_manual
precision_mush_manual = cm_mush_tp_manual / (cm_mush_tp_manual + cm_mush_fp_manual)
recall_mush_manual = cm_mush_tp_manual / (cm_mush_tp_manual + cm_mush_fn_manual)
f1_score_mush_manual = 2 * (precision_mush_manual * recall_mush_manual)/(precision_mush_manual + recall_mush_manual)

print('For manual, ERR: ' + str(error_mush_manual) + ', ACC: ' + str(accuracy_mush_manual) + ', F1 Score: ' + str(f1_score_mush_manual))
```
