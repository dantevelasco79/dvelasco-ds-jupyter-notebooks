## 1a)
#### theta(0) = SampMean(Y) - theta(1) * SampMean(X)
#### theta(1) = Cov(X,Y) / Var(X)
#### Cov(X,Y) = Corr(X,Y) * StdDev(X) * StdDev(Y)\
#### Var(X) = StdDev(X)^2
Var(X) = 10^2 = 100

Cov(X,Y) = .8 * 10 * 20 = 160

theta(1) = 160/100 = 1.6

theta(0) = 100 - 1.6 * 50 = 100 - 80 = 20

#### theta(0) = 20
#### theta(1) = 1.6

## 1b)
Var(X) = 10^2 = 100

Cov(X,Y) = -.8 * 10 * 20 = -160

theta(1) = -160/100 = -1.6

theta(0) = 100 - -1.6 * 50 = 100 + 80 = 180

theta(0) = 180
theta(1) = -1.6

#### The absolute value of all parameters but theta(0) stay the same. However, because theta(1) is now negative, just as expected based on the corr it is now a downward slope, meaning that as x goes up y goes down and vice versa.

## 1c)

We know that theta(0) or Intercept = SampMean(Y) - Slope * SampMean(X)
We can also dervive that Slope = (SampMean(Y) - Intercept) / SampMean(X)
Lastly, we can dervive that SampMean(Y) = Slope * SampMean(X) + Intercept

#### Therefore, both the Slope and Intercept are being formed based on the connection between the two sample means for X and Y. As a Slope formed from two points must always pass through those two points, our model must always pass through the sample means of x and y.

```python
import numpy as np
from numpy.linalg import inv

clf_theta = np.matmul(np.matmul(inv(np.matmul(np.transpose(X_train.values), X_train.values)), np.transpose(X_train.values)), y_train.values)

def pred(x_values, theta):
    pred = []
    for idx in range(len(x_values)):
        sum = 0
        for jdx in range(len(x_values[idx])):
            sum += x_values[idx][jdx] * theta[jdx]
        pred.append(sum)
    return pred
```

```python
import random as rd
from numpy import linalg as LA

def grad_desc(alpha, max_iter, init_range, min_change):
    iter = 0
    theta = []
    for idx in range(len(X_train.values[0])):
        theta.append(rd.randint(-init_range, init_range))
    min_theta = theta
    while(iter < max_iter):
        iter += 1
        change = (alpha * (2 / len(y_train.values))) * np.matmul(np.transpose(np.matmul(X_train.values, theta) - y_train.values), X_train.values)
        if (LA.norm(theta - change - theta) < min_change):
            if (mean_squared_error(y_train.values, pred(X_train.values, theta)) < mean_squared_error(y_train.values, pred(X_train.values, min_theta))):
                min_theta = theta
            for idx in range(len(X_train.values[0])):
                theta[idx] = rd.randint(-init_range, init_range)
        else:
            theta = theta - change
    if (mean_squared_error(y_train.values, pred(X_train.values, theta)) < mean_squared_error(y_train.values, pred(X_train.values, min_theta))):
        min_theta = theta
    return min_theta
```

## 5b)

```python
dict = {'alpha': [], 'max_iterations': [], 'MSE_train': [], 'R2_train': [], 'MSE_test': [], 'R2_test': []}
alphas = [0.01, 0.1, 0.5]
iters = [10, 50, 100]

for i in range(3):
    for j in range(3):
    
        dict['alpha'].append(alphas[i])
        dict['max_iterations'].append(iters[j])
        
        grad_theta = grad_desc(alphas[i], iters[j], 100, .01)
    
        y_train_grad_predicted = pred(X_train.values, grad_theta) 

        MSE_grad_train = mean_squared_error(y_train.values, y_train_grad_predicted)
        R2_grad_train = r2_score(y_train.values, y_train_grad_predicted)
    
        dict['MSE_train'].append(MSE_grad_train)
        dict['R2_train'].append(R2_grad_train)
    
        y_test_grad_predicted = pred(X_test.values, grad_theta) 

        MSE_grad_test = mean_squared_error(y_test.values, y_test_grad_predicted)
        R2_grad_test = r2_score(y_test.values, y_test_grad_predicted)

        dict['MSE_test'].append(MSE_grad_test)
        dict['R2_test'].append(R2_grad_test)

grad_df = pd.DataFrame(dict)
grad_df
```

## 6a)

J(theta) = 1/N ||X(theta) - y||^2 + lambda(1/d) ||theta||^2

der(J(theta)) = 0

0 = (X(theta) - y)^T * X + (lambda(theta))

theta(X^T * X(theta) + lambda * I) = X^T * y

#### theta = (X^T * X(theta) + lambda * I)^(-1) * X^T * y

## 6b)

```python
def mse_ridge(lambda_ridge, theta, y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    sum = 0
    for idx in range(len(theta)):
        sum += theta[idx]**2
    sum = sum * lambda_ridge
    return mse + sum

def grad_desc_ridge(alpha, max_iter, init_range, min_change, lambda_ridge, x_values, y_values):
    iter = 0
    theta = []
    for idx in range(len(x_values[0])):
        theta.append(rd.randint(-init_range, init_range))
    min_theta = theta
    while(iter < max_iter):
        iter += 1
        change = (alpha * (2 / len(y_values))) * np.matmul(np.transpose(np.matmul(x_values, theta) - y_values), x_values)
        new_theta = []
        for idx in range(len(theta)):
            new_theta.append(theta[idx] * (1 - alpha * lambda_ridge))
        new_theta = new_theta - change
        if (LA.norm(new_theta - theta) < min_change):
            if (mse_ridge(lambda_ridge, theta, y_values, pred(x_values, theta)) < mse_ridge(lambda_ridge, min_theta, y_values, pred(x_values, min_theta))):
                min_theta = theta
            for idx in range(len(x_values[0])):
                theta[idx] = rd.randint(-init_range, init_range)
        else:
            theta = new_theta
    if (mse_ridge(lambda_ridge, theta, y_values, pred(x_values, theta)) < mse_ridge(lambda_ridge, min_theta, y_values, pred(x_values, min_theta))):
        min_theta = theta
    return min_theta
```
