```python
import pandas

df = pandas.read_csv('kc_house_data.csv')
df
```

```python
means = []
variances = []

columns = df.columns.tolist()
columns.remove('id')
columns.remove('date')
columns.remove('zipcode')
columns
```

```python
for column in columns:
    means.append(df[column].mean())
    variances.append(df[column].var())
    print('feature: ' + str(column))
    print('mean: ' + str(df[column].mean()))
    print('min: ' + str(df[column].min()))
    print('max: ' + str(df[column].max()))
    print('variance: ' + str(df[column].var()) + '\n')
```

```python
print('The feature with the lowest average is longitude with a mean of ' + str(min(means)))
print('The feature with the greatest average is price with a mean of ' + str(max(means)))
print('The feature with the lowest variance is waterfront with a mean of ' + str(min(variances)))
print('The feature with the greatest variance is price with a mean of ' + str(max(variances)))
```

```python
dict = {'name': [], 'corr': []}
for column in columns:
    dict['name'].append(column)
    dict['corr'].append(df[column].corr(df['price']))
corr_df = pandas.DataFrame(dict)
corr_df
```

Every feature is positiively correlated to the response. The feature with the greatest correlation to price besides itself is sqft_living. I was unable to find a negative correlation coefficient with response.
